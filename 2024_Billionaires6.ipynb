# DATA CLEANING AND MANIPULATION PROCESS

```python
data['Rank'] = pd.Series(data.index).rank().astype(int)
```

```python
data.rename(columns={'Citizenship':'country'},inplace=True)
data.rename(columns={'2024 Net Worth	':'net worth'},inplace=True)
data.rename(columns={'Source of Wealth':'source'},inplace=True)
data.rename(columns={'Age':'age'},inplace=True)
data.rename(columns={'Gender':'gender'},inplace=True)
```

```python
data['age']=data['age'].fillna(0).astype(int)
```

```python
data.rename(columns={'2024 Net Worth':'net worth'},inplace=True)
```

```python
data['age']=data['age'].fillna(0).astype(int)
```

```python
data['net worth'] = data['net worth'].str.replace('$', '', regex=False).str.replace('B', '', regex=False).str.replace('M', '', regex=False)
data['net worth']=data['net worth'].fillna(0).astype(float)
data['net worth']=data['net worth'].fillna(0).astype(int)
```

```python
data
```

```python
# data.drop(columns=['Marital Status'],inplace=True)
data.drop(columns=['Self-Made'],inplace=True)
data.drop(columns=['Self-Made Score'],inplace=True)
data.drop(columns=['Philanthropy Score'],inplace=True)
data.drop(columns=['Residence'],inplace=True)
data.drop(columns=['Children'],inplace=True)
data.drop(columns=['Education'],inplace=True)
```

```python
columns=['Rank','Name','net worth','source','Industry','Organization','country','age','gender']
data=data[columns]
data
```
